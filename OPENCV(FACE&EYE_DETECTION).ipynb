```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
face_cascade=cv.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face_cascade
```

```python
%matplotlib inline
img = cv.imread('images.jpeg')
```

```python
plt.imshow(img)
```

```python
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
faces = face_cascade.detectMultiScale(image,1.1,4,minSize=(30,30))
```

```python
faces
```

```python
len(faces)
```

```python
for (x,y,w,h) in faces:
    cv.rectangle(image,(x-5,y-5),(x+w+5,y+h+5),color=(0,255,0),thickness=2)
```

```python
plt.imshow(image)
```

```python
cap = cv.VideoCapture('vdo.mp4')
```

```python
while True:
    ret,frame = cap.read()
    grey = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(grey ,1.1 ,4 ,minSize=(30,30))

    for(x,y,w,h) in faces:
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    
    
    image = cv.resize(frame, (200, 200))
    cv.imshow("video",image)
    if cv.waitKey(1)& 0xFF ==27:
        break
cap.release()
cv.destroyAllWindows()
```
